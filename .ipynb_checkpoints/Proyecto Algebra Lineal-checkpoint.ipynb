## 0. Conceptos
Para poder modificar las funciones solicitadas es necesario que comprenda algunos conceptos, por lo cual es necesario que defina:

1. ¿En que consiste la convolución y cómo puede ser calculada? (Puede dar una definición matemática y/o ejemplos).
    -La mayoría de filtros de imagenes que desenfocan, realzan bordes y destacan bordes, entre otras, utiliza una              convolución. 
    La matriz de convolución o kernel, puede tomar cualquier tamaño depende del filtro que se desee aplicar. 
    La convolución se define matemáticamente para modelos  bidimensionales  a  partir  de  la  ecuación:
    g(x, y) = f(x, y)⋅K(x, y)
      = ∫∫f(i, j)⋅K(x-i, y-j) en(-∞) y ∞ 
      = ∑ ∑f(i,y)⋅K(x-i, y-j) de 0 a ∞.
    una convulción trabaja en el radio del pixel (x, y) 

2. ¿Qué es el *kernel* de un filtro y para que se utiliza?
    -Se entiende por kernel a la matriz de coeficientes  de la imagen donde el entorno del pixel (x, y) de la imagen para obtener otro valor g(x,y) está determinado por el tamaño y por la forma del kernel. La mayor parte de las veces se seleciona un kernel cuadrado. 
    
3. ¿En qué consiste un filtro Gaussiano?
    - Es un filtro que consiste en emborronar y eliminar el ruido de las imágenes. 
    Se modela con la función: 
    G(x, y) = [1 / (2πσ^2)]^[-(x^2 + y^2)/σ^2]
    Tiene un kernel de 5x5 y la suma de los valores de la matriz da como resultado 273
                       [[ 1  4  7  4  1]
                        [ 4 16 26 16  4]
                        [ 7 26 41 26  7]
                        [ 4 16 26 16  4]
                        [ 1  4  7  4  1]]
4. ¿En qué consiste el Filtro Sobel? ¿para qué sirve?
  - El filtro sobel detecta los bordes horizontales y verticales de una imagen a escala de grises. 
  Con siste en un kernel de 3x3 y la suma de todos los valores de la matriz da cero como resltado. 
  Sobe 
       x = [[-1  0  1]
            [-2  0  2]
            [-1  0  2]]
            
   Sobel 
         y = [[-1 -2 -1]
              [ 0  0  0]
              [ 1  2  1]]

3. Describa el algoritmo de detección de bordes Canny.
    Si f(x,y) es la intensidad de un  pixel que se encuentra en la posición (x, y) de la imagen.
    a. Se aplica el filtro Gaussiano para suavizar la imagen. 
           G(x, y) = [1 / (2πσ^2)]^[-(x^2 + y^2)/σ^2]
    b. Se calcula el gradiente de g para maximizar la diferencia entre el ruido y la señal. 
    c. Se calcula el modulo y el argumento de la imagen suavizada y filtrada. 
    d. En la dirección del gradiente se eliminan aquellos pixeles que no sean de máximos o mínimos locales del módulo. Ya que el operador gradiente aplicado a un pixel (x,y) devuelve un vector que indica la dirección máxima variabilidad de la intensidad luminosa y su nivel de variación. 
       e.Se construyen dos imágenes binarias auxiliares: I1 e I2 a partir de dos umbrales T1 < T2, tales que, con T1 habrían muchos mas píxeles bordes que con T2. 
    f.  a partir de píxeles bordes de la imagen I1, se van construyendo bordes, mediante la adición de píxeles bordes de la imagen I2.
    g. Se calcula la media (Mb) de los píxeles que definitivamente fueron bordes.
    h.Si f(x,y)>Mb, entonces el píxel (x,y) pertenece a la señal, si no pertenece al fondo

````python
import numpy as np
import matplotlib.pylab as plt
import cv2
import time

from moviepy.editor import VideoFileClip
from IPython.display import HTML

%matplotlib inline
````

````python
def multiplicacionSobelYKernel(A, B):
    kernel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    valor = 0
    for x in range (0, 3):
        for y in range (0, 3):
            valor += (A[x, y]*kernel[x, y])
    B[2,2] = valor/273

    
def multiplicacionGaussianKernel(A, B):
    kernel = np.array([[1,4,7,4,1],[4,16,26,16,4],[7,26,41,26,7],[4,16,26,16,4],[1,4,7,4,1]])
    valor = 0
    for x in range (0, 5):
        for y in range (0, 5):
            valor += (A[x, y]*kernel[x, y])
    B[2,2] = valor/273

def multiplicacionSobelXKernel(A, B):
    kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,2]])
    valor = 0
    for x in range (0, 3):
        for y in range (0, 3):
            valor += (A[x, y]*kernel[x, y])
    B[2,2] = valor/273  

def multiplicacionKernel(A, kernel):
    B = np.array(A)
    if(kernel == "sx"):
        for x in range (0, len(A)-2):
            for y in range (0, len(A[0])-2):
                multiplicacionSobelXKernel(A[x:x+3, y:y+3], B[x:x+3, y:y+3])
    if(kernel == "sy"):
        for x in range (0, len(A)-2):
            for y in range (0, len(A[0])-2):
                multiplicacionSobelYKernel(A[x:x+3, y:y+3], B[x:x+3, y:y+3])
    if(kernel == "g"):
         for x in range (0, len(A)-5):
            for y in range (0, len(A[0])-5):
                multiplicacionGaussianKernel(A[x:x+5, y:y+5], B[x:x+5, y:y+5])
    return B
````

````python
def gaussian_blur(img, kernel = 5):
    '''Modificar esta función'''
    return multiplicacionKernel(img, "sx")

def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    # Calculate directional gradient
    if orient == 'x':
        x,y = 1,0
    elif orient == 'y':
        x,y = 0,1
        
    '''Modificar esta función'''
    sobel = multiplicacionKernel(img, "s")
    
    # Take the absolute value of the derivative or gradient
    abs_sobel = np.absolute(sobel)

    # Scale to 8-bit (0 - 255) then convert to type = np.uint8
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    
    # Create a mask of 1's where the scaled gradient magnitude 
    # is > thresh_min and < thresh_max
    sbinary = np.zeros_like(scaled_sobel)
    sbinary[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    return sbinary

def mag_thresh(image, sobel_kernel=3, mag_thresh=(0, 255)):
    # Calculate gradient magnitude
    # Take the gradient in x and y separately
    sobelx = multiplicacionKernel(image, "sx")
    sobely = multiplicacionKernel(image, "sy")
    
    # Calculate the magnitude 
    abs_sobel = np.sqrt(sobelx**2 +sobely**2)

    # Scale to 8-bit (0 - 255) and convert to type = np.uint8
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    
    # Create a binary mask where mag thresholds are met
    sbinary = np.zeros_like(scaled_sobel)
    sbinary[(scaled_sobel >= mag_thresh[0]) & (scaled_sobel <= mag_thresh[1])] = 1
    return sbinary

def dir_threshold(image, sobel_kernel=3, thresh=(0, np.pi/2)):
    # Calculate gradient direction
    # Take the gradient in x and y separately
    sobelx = multiplicacionKernel(image, "sx")
    sobely = multiplicacionKernel(image, "sy")

    # Take the absolute value of the x and y gradients
    abs_sobelx = np.absolute(sobelx)   
    abs_sobely = np.absolute(sobely)
    
    # Use np.arctan2(abs_sobely, abs_sobelx) to calculate the direction of the gradient 
    grad_dir = np.arctan2(abs_sobely, abs_sobelx)
    
    # Create a binary mask where direction thresholds are met
    sbinary = np.zeros_like(grad_dir)
    sbinary[(grad_dir >= thresh[0]) & (grad_dir <= thresh[1])] = 1
    return sbinary
````

## 1. Prueba en Imagen

````python
start_time = time.time()
# Test filter on single image
im = plt.imread("BTD.jpg")
gray = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)

# Gaussian Blur Filtering
gray = gaussian_blur(gray, kernel = 5)

# Choose a Sobel kernel size
ksize = 3 # Choose a larger odd number to smooth gradient measurements

# Apply each of the thresholding functions
gradx = abs_sobel_thresh(gray, orient='x', sobel_kernel=ksize, thresh=(50, 100))
grady = abs_sobel_thresh(gray, orient='y', sobel_kernel=ksize, thresh=(50, 100))
mag_binary = mag_thresh(gray, sobel_kernel=ksize, mag_thresh=(20, 100))
dir_binary = dir_threshold(gray, sobel_kernel=ksize, thresh=(0, np.pi/2))

# Create combined image with sobel filters applied
combined = np.zeros_like(dir_binary)
combined[((gradx == 1) & (grady == 1)) | ((mag_binary == 1) & (dir_binary == 1))] = 1

rgb_combined = np.asarray(np.dstack((combined, combined, combined)), dtype=np.float64)
````

## 2. Prueba en Videos

````python
def process_image(image):
    
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    gray = gaussian_blur(gray, kernel = 5)
    # Choose a Sobel kernel size
    ksize = 3 # Choose a larger odd number to smooth gradient measurements

    # Apply each of the thresholding functions
    gradx = abs_sobel_thresh(gray, orient='x', sobel_kernel=ksize, thresh=(50, 100))
    grady = abs_sobel_thresh(gray, orient='y', sobel_kernel=ksize, thresh=(50, 100))
    mag_binary = mag_thresh(gray, sobel_kernel=ksize, mag_thresh=(20, 100))
    dir_binary = dir_threshold(gray, sobel_kernel=ksize, thresh=(0, np.pi/4))
    
    combined = np.zeros_like(dir_binary)
    combined[((gradx == 1) & (grady == 1)) | ((mag_binary == 1) & (dir_binary == 1))] = 1
    
    
    rgb_combined = np.asarray(np.dstack((combined, combined, combined)), dtype=np.uint8)
    
    retval, threshold = cv2.threshold(rgb_combined, 0, 255, cv2.THRESH_BINARY)
    
    return threshold
````

````python
filter_output = 'Batman_Filter.mp4'
clip1 = VideoFileClip("Batman.mp4")
batman_clip = clip1.fl_image(process_image) #NOTE: this function expects color images!!
%time batman_clip.write_videofile(filter_output, audio=False)
````

````python
cap = cv2.VideoCapture(0)  # 0 for Python 3, 1 for Python 2. No idea why?!!

out = cv2.VideoWriter("output.avi", cv2.VideoWriter_fourcc(*"MJPG"), 20,(640,480))

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        frame = cv2.flip(frame,180)
        new_frame = process_image(frame)
        out.write(new_frame)

        cv2.imshow('frame',new_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
````
